## Train Split Validation

```python
seed = 42
np.random.seed(seed)
prop_vec = [14, 3, 3]
train_x, train_y, test_x, test_y, val_x, val_y = utils.train_test_val_split(all_x, all_y, prop_vec, shuffle=True, seed=seed-1)
```

## SVM

```python
# not using this one since accuracy isn't high enough

from sklearn.svm import SVC, LinearSVC
from sklearn.base import clone

svm = SVC(kernel='linear', random_state=seed)

svm.fit(train_x, train_y)
```

```python
def model_accuracy(model, x, true_y):
    pred = model.predict(x)
    return np.sum(pred == true_y) / true_y.shape[0]

name = 'svm'
train_acc = model_accuracy(svm, train_x, train_y)
val_acc = model_accuracy(svm, val_x, val_y)

print('[{}] Train accuracy: {:.2f}%, Val accuracy: {:.2f}%'.format(name, train_acc*100, val_acc*100))
```

## RainForestClassifier

```python
# giving the self-reported values 0-3 as label 0 (no anxiety) and 4-7 as label 1 (anxiety)

d_map = {0:0,
         1:0,
         2:0,
         3:0,
         4:1,
         5:1,
         6:1,
         7:1}
df['TIPI4'] = df['TIPI4'].map(d_map)
df.head()
```

```python
X = df.drop(columns='TIPI4')
y = df['TIPI4']
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(X,y, test_size=0.3,random_state=10)

from sklearn.impute import SimpleImputer

fill_values = SimpleImputer(missing_values=0, strategy="mean")

train_x = fill_values.fit_transform(train_x)
test_x = fill_values.fit_transform(test_x)


from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(random_state=10)

random_forest_model.fit(train_x, train_y.ravel())
RandomForestClassifier(random_state=10)

name = 'rain forest classifier'
predict_train_data = random_forest_model.predict(test_x)

from sklearn import metrics

score = metrics.accuracy_score(test_y, predict_train_data)
score = score * 100

print("[{}] Accuracy: {:.2f}%".format(name, score))
```
